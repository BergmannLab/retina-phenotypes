## Most common genes:

```python
df_genes_count.head(5)
df_genes_count_reduced= df_genes_count.head(50)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.xticks(rotation=90)
ax.bar(df_genes_count_reduced['Unnamed: 0'], df_genes_count_reduced['gen'])
plt.title('Most common genes')
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
plt.xticks(rotation=90)
ax.bar(df_genes_count_reduced['Unnamed: 0'], df_genes_count_reduced['ratio_N_pheno'])
plt.title('Most common genes ratio')
```
