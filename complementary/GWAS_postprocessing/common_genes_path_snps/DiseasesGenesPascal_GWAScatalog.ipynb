## 1- Read the Neale Pascal genes for all the diseases

```python
p_value_min = 5.7  # -math.log10(0.05/len(x))

#input_dir='/HDD/data/ukbb/disease_sumstats/files_modified/'
input_dir='/HDD/data/ukbb/disease_sumstats/VARIANTS/'
l_diseases_all=[]

#os.chdir('/HDD/data/ukbb/disease_sumstats/files_modified/')
os.chdir('/HDD/data/ukbb/disease_sumstats/VARIANTS/')
for file in glob.glob("*__gene_scores"):
    #print(file)
    #file='4080_irnt.gwas.imputed_v3.both_sexes.tsv'
    #df = pd.read_csv(input_dir+file+'__gene_scores', delimiter='\t', names =['gen', 'p']) #, index_col=None, header=0)
    df = pd.read_csv(input_dir+file, delimiter='\t', names =['gen', 'p']) #, index_col=None, header=0)

    df['file_col']=file

    df['-log10(p)'] = -np.log10(df['p'])
    y = df[df['-log10(p)'] >= p_value_min]
    df_significant = y.sort_values('-log10(p)', ascending=False)
    pheno, b = file.split(".gwas.imputed_v3.both_sexes.tsv__gene_scores")
    aux, pheno_name = pheno.split("only_rs_notna_mod_variants_")
    
    data={
    'pheno':  pheno,
    'name_pheno': pheno_name,
    'N genes': len(df_significant),
    'names': df_significant['gen'].to_list()
    }
    l_diseases_all.append(data)

df_diseases_all =pd.DataFrame(l_diseases_all)
    
    #print(pheno, ' Significant number of genes: ', len(df_significant))
```

#### b) GWAS catalog

```python
dir_catalog = '/HDD/data/GWAS_catalog/'
os.chdir(dir_catalog)

l_genes_shared=[]
for file_name in glob.glob("gwas-association-downloaded_SBP.tsv"): #glob("*.tsv"):
    #print(file_name)
    df_gwas = pd.read_table(dir_catalog+ str(file_name))
    #df_reduced = df_gwas[['MAPPED_TRAIT', 'REPORTED GENE(S)', 'MAPPED_GENE']]
    #print(df_reduced.head(5))
    for i in range(len(df_gwas)):
        gen = df_gwas['REPORTED GENE(S)'].iloc[i]
        for j in range(len(df_genes_neale_red)):
            for m in range(len(df_genes_neale_red['names'].iloc[j])):
                if df_genes_neale_red['names'].iloc[j][m]==gen:
                    data={
                    'Neale_pheno':  df_genes_neale_red['name_pheno'].iloc[j],
                    'MAPPED_TRAIT': df_gwas['MAPPED_TRAIT'].iloc[i],
                    'gene': df_genes_neale_red['names'].iloc[j][m],
                    'gene1': gen,
                    'INITIAL SAMPLE SIZE': df_gwas['INITIAL SAMPLE SIZE'].iloc[i]
                    }
                    l_genes_shared.append(data)

df_genes_shared =pd.DataFrame(l_genes_shared)
```
